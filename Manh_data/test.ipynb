```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('Razer DeathAdder Essential Gaming Mouse.csv')
```

```python
# xử lí

parts = df['country&time'].str.split(pat='in | on',expand=True, n = 2)
df['country'] = parts[1]
df['time'] = parts[2]
df = df.drop(columns=['country&time'])
```

```python
def convert_date(date_string):
    date_string = date_string.strip()
    dt_object = datetime.datetime.strptime(date_string, "%B %d, %Y")
    return dt_object.strftime("%d/%m/%Y")
```

```python
df['time'] = df['time'].apply(convert_date)
# sắp xếp lại các cột
columns_order = ['user', 'star', 'time', 'kind item', 'sumary', 'comment', 'image', 'country']
df = df.reindex(columns=columns_order)
df
```
