```python
from bs4 import BeautifulSoup
import urllib.request
import csv
import time
```

```python
def text2date(text):
    text_split = text.split()
    year = text_split[-1]
    day = text_split[-2][:-1]
    month = text_split[-3]
    
    M = {'January'   : 1,
         'February'  : 2,
         'March'     : 3,
         'April'     : 4,
         'May'       : 5,
         'June'      : 6,
         'July'      : 7,
         'August'    : 8,
         'September' : 9,
         'October'   : 10,
         'November'  : 11,
         'December'  : 12}
    
    return f'{int(day):02d}/{M[month]:02d}/{year}'
```

```python
fields = ['User', 'Number of star', 'Date', 'Type of Product', 'Review Title', 'Review Content', 'Image in Review ']
filename = "test.csv"
with open(filename, 'w') as csvfile:
    csvwriter = csv.writer(csvfile)

    csvwriter.writerow(fields)
```

```python
type_star = ['five']
sample = 834
for star in type_star:
    print('\n' + star)
    if sample == 1000:
        break
    next_star = False
    page = 1
    while not next_star:
        try:
            url = f'https://www.amazon.com/product-reviews/B093BSR3H4/ref=cm_cr_arp_d_paging_btm_next_{page}?ie=UTF8&filterByStar=five_star&reviewerType=all_reviews&pageNumber={page}#reviews-filter-bar'
            web = urllib.request.Request(url, headers={'User-Agent': 'Mozilla/5.0'})
            web = urllib.request.urlopen(web)
            soup = BeautifulSoup(web, 'html.parser')

            if soup.find('div', {'class' : 'a-section a-spacing-top-large a-text-center no-reviews-section'}):
                next_star = True
            
            print(f'\nPage {page}:')
            rows = []
            reviews = soup.find_all('div', {'data-hook' : 'review'})
            for review in reviews:
                sleep = 1
                user = review.find('div', {'data-hook' : 'genome-widget'}).text

                n_star = review.find('i', {'data-hook' : 'review-star-rating'}).text
                number_of_star = n_star[0]

                title = review.find('a', {'data-hook' : 'review-title'}).text.replace(n_star, '').strip()

                date = text2date(review.find('span', {'data-hook' : 'review-date'}).text)

                type = review.find('a', {'data-hook' : 'format-strip'}).text.replace('Size: ', '').replace('Style: ', ' - ')

                content = review.find('span', {'data-hook' : 'review-body'}).text.strip()

                image = 1 if review.find('div', {'class' : 'a-section a-spacing-medium review-image-container'}) == None else 0

                rows.append([user, number_of_star, date, type, title, content, image])
                sample += 1
                print(sample, end=' ')
            
            if rows:
                filename = "test.csv"
                with open(filename, 'a') as csvfile:
                    csvwriter = csv.writer(csvfile)
                    csvwriter.writerows(rows)
                    page += 1
            time.sleep(10)
                
        except Exception as error:
            if '404' in str(error):
                print('System is sleeping!')
                time.sleep(sleep * 300)
                print(f'System has slept in {sleep * 300}!')
                sleep += 1
            else:
                print(error)
```
